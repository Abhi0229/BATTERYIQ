```python
import os
import pandas as pd
import numpy as np

# ✅ Simple relative path — you're already in cleaned_dataset folder
data_path = "data"

# ✅ Get all CSV files sorted
all_files = sorted([f for f in os.listdir(data_path) if f.endswith('.csv')])

print(f"✅ Total cycle files found: {len(all_files)}")
print(f"🔢 First 5: {all_files[:5]}")
print(f"🔢 Last  5: {all_files[-5:]}")
```

```python
# ✅ Build Master DataFrame — one row per cycle
records = []

for filename in all_files:
    filepath = os.path.join(data_path, filename)
    
    try:
        df = pd.read_csv(filepath)
        
        # Skip files that are empty or too small
        if df.shape[0] < 10:
            continue
        
        # Extract cycle number from filename
        cycle_number = int(filename.replace('.csv', ''))
        
        record = {
            'cycle'                : cycle_number,
            'voltage_mean'         : df['Voltage_measured'].mean(),
            'voltage_min'          : df['Voltage_measured'].min(),
            'voltage_std'          : df['Voltage_measured'].std(),
            'current_mean'         : df['Current_measured'].mean(),
            'current_std'          : df['Current_measured'].std(),
            'current_min'          : df['Current_measured'].min(),
            'temp_mean'            : df['Temperature_measured'].mean(),
            'temp_max'             : df['Temperature_measured'].max(),
            'temp_std'             : df['Temperature_measured'].std(),
            'voltage_load_mean'    : df['Voltage_load'].mean(),
            'voltage_load_min'     : df['Voltage_load'].min(),
            'current_load_mean'    : df['Current_load'].mean(),
            'duration'             : df['Time'].max() - df['Time'].min(),
            'capacity'             : np.trapz(
                                        np.abs(df['Current_measured']),
                                        df['Time']
                                     ) / 3600
        }
        
        records.append(record)
    
    except Exception as e:
        print(f"⚠️ Skipped {filename}: {e}")

# ✅ Convert to DataFrame
master_df = pd.DataFrame(records)

print(f"✅ Master DataFrame shape: {master_df.shape}")
print(f"\n📋 Columns:\n{master_df.columns.tolist()}")
print(f"\n🔍 First 3 rows:")
print(master_df.head(3))
```

```python
import pandas as pd
import numpy as np
import os

# ✅ Load metadata
meta = pd.read_csv("metadata.csv")

# ✅ Filter ONLY discharge cycles (that's where battery is being USED)
discharge = meta[meta['type'] == 'discharge'].copy()

print(f"✅ Total cycles       : {len(meta)}")
print(f"✅ Discharge cycles   : {len(discharge)}")
print(f"✅ Unique batteries   : {meta['battery_id'].unique()}")
print(f"\n🔍 Discharge sample:")
print(discharge[['battery_id', 'filename', 'Capacity', 'ambient_temperature']].head(10))
```

```python
# ✅ Build master DataFrame by merging metadata + CSV sensor features
data_path = "data"
records = []

for _, row in discharge.iterrows():
    filepath = os.path.join(data_path, row['filename'])
    
    try:
        df = pd.read_csv(filepath)
        
        # Skip if key columns missing
        required_cols = ['Voltage_measured', 'Current_measured', 'Temperature_measured', 'Time']
        if not all(col in df.columns for col in required_cols):
            continue
            
        # Skip if too small
        if df.shape[0] < 10:
            continue
        
        record = {
            # --- From metadata ---
            'battery_id'           : row['battery_id'],
            'filename'             : row['filename'],
            'capacity'             : float(row['Capacity']),
            'ambient_temperature'  : row['ambient_temperature'],
            
            # --- From CSV sensor readings ---
            'voltage_mean'         : df['Voltage_measured'].mean(),
            'voltage_min'          : df['Voltage_measured'].min(),
            'voltage_std'          : df['Voltage_measured'].std(),
            'current_mean'         : df['Current_measured'].mean(),
            'current_std'          : df['Current_measured'].std(),
            'current_min'          : df['Current_measured'].min(),
            'temp_mean'            : df['Temperature_measured'].mean(),
            'temp_max'             : df['Temperature_measured'].max(),
            'temp_std'             : df['Temperature_measured'].std(),
            'duration'             : df['Time'].max() - df['Time'].min(),
        }
        
        # Add Voltage_load if available
        if 'Voltage_load' in df.columns:
            record['voltage_load_mean'] = df['Voltage_load'].mean()
            record['voltage_load_min']  = df['Voltage_load'].min()
        else:
            record['voltage_load_mean'] = np.nan
            record['voltage_load_min']  = np.nan
            
        records.append(record)
        
    except Exception as e:
        print(f"⚠️ Skipped {row['filename']}: {e}")

# ✅ Convert to DataFrame
master_df = pd.DataFrame(records)

print(f"✅ Master DataFrame shape: {master_df.shape}")
print(f"\n📋 Columns: {master_df.columns.tolist()}")
print(f"\n🔍 First 5 rows:")
print(master_df.head())
print(f"\n📊 Capacity stats:")
print(master_df['capacity'].describe())
```

```python
# Load clean data
master_df = pd.read_csv("master_df_clean.csv")

print("⚙️ Engineering new features...")

# ============================================
# FEATURE 1: cycle_number per battery
# Each battery starts its own count from 1
# ============================================
master_df = master_df.sort_values(['battery_id', 'filename']).reset_index(drop=True)
master_df['cycle_number'] = master_df.groupby('battery_id').cumcount() + 1

# ============================================
# FEATURE 2: Initial capacity per battery
# (capacity at cycle 1 for that battery)
# ============================================
initial_cap = master_df.groupby('battery_id')['capacity'].transform('first')
master_df['initial_capacity'] = initial_cap

# ============================================
# FEATURE 3: Capacity drop & % drop
# How much has it degraded from its fresh state?
# ============================================
master_df['capacity_drop']     = master_df['initial_capacity'] - master_df['capacity']
master_df['capacity_drop_pct'] = (master_df['capacity_drop'] / master_df['initial_capacity']) * 100

# ============================================
# FEATURE 4: Voltage range
# High range = unstable battery
# ============================================
master_df['voltage_range'] = master_df['voltage_mean'] - master_df['voltage_min']

# ============================================
# FEATURE 5: Temperature rise
# How much did temp spike above average?
# ============================================
master_df['temp_rise'] = master_df['temp_max'] - master_df['temp_mean']

# ============================================
# FEATURE 6: Power mean (Watts)
# Power = Voltage × Current
# ============================================
master_df['power_mean'] = master_df['voltage_mean'] * abs(master_df['current_mean'])

# ============================================
# FEATURE 7: Energy delivered (Wh)
# Energy = Power × Time
# ============================================
master_df['energy_delivered'] = master_df['power_mean'] * master_df['duration'] / 3600

# ============================================
# FEATURE 8: Efficiency proxy
# How much energy per unit of capacity?
# ============================================
master_df['efficiency_proxy'] = master_df['energy_delivered'] / master_df['capacity']

# ============================================
# PREVIEW RESULTS
# ============================================
print(f"✅ New shape: {master_df.shape}")
print(f"\n📋 All columns ({len(master_df.columns)}):")
for col in master_df.columns:
    print(f"   → {col}")

print(f"\n🔍 New features sample:")
new_cols = ['battery_id', 'cycle_number', 'capacity', 'capacity_drop_pct',
            'voltage_range', 'temp_rise', 'power_mean', 'energy_delivered', 'efficiency_proxy']
print(master_df[new_cols].head(8).round(3))
```

```python
import pandas as pd
import numpy as np

master_df = pd.read_csv("master_df_featured.csv")

# ============================================
# TARGET 1: stress_label
# Based on capacity degradation percentage
# ============================================
def assign_stress(drop_pct):
    if drop_pct < 10:
        return 'Low'
    elif drop_pct < 25:
        return 'Medium'
    else:
        return 'High'

master_df['stress_label'] = master_df['capacity_drop_pct'].apply(assign_stress)

print("✅ TARGET 1: stress_label")
print(master_df['stress_label'].value_counts())
print(f"   Distribution: {master_df['stress_label'].value_counts(normalize=True).round(2).to_dict()}")

# ============================================
# TARGET 2: health_score (0-100)
# How healthy is the battery right now?
# ============================================
master_df['health_score'] = (master_df['capacity'] / master_df['initial_capacity']) * 100
master_df['health_score'] = master_df['health_score'].clip(0, 100).round(2)

print("\n✅ TARGET 2: health_score")
print(f"   Min : {master_df['health_score'].min():.2f}")
print(f"   Mean: {master_df['health_score'].mean():.2f}")
print(f"   Max : {master_df['health_score'].max():.2f}")

# ============================================
# TARGET 3: RUL (Remaining Useful Life)
# End of life = capacity drops below 70% of initial
# ============================================
rul_list = []

for battery_id, group in master_df.groupby('battery_id'):
    group = group.sort_values('cycle_number').reset_index(drop=True)
    
    # Find the threshold: 70% of this battery's initial capacity
    eol_threshold = group['initial_capacity'].iloc[0] * 0.70
    
    # Find the last cycle where capacity is still above threshold
    above_threshold = group[group['capacity'] >= eol_threshold]
    
    if len(above_threshold) > 0:
        eol_cycle = above_threshold['cycle_number'].max()
    else:
        eol_cycle = group['cycle_number'].min()
    
    # RUL = cycles remaining until end of life
    group['RUL'] = (eol_cycle - group['cycle_number']).clip(lower=0)
    rul_list.append(group[['filename', 'RUL']])

rul_df = pd.concat(rul_list).reset_index(drop=True)
master_df = master_df.merge(rul_df, on='filename', how='left')

print("\n✅ TARGET 3: RUL")
print(f"   Min : {master_df['RUL'].min():.0f} cycles")
print(f"   Mean: {master_df['RUL'].mean():.0f} cycles")
print(f"   Max : {master_df['RUL'].max():.0f} cycles")

# ============================================
# TARGET 4: efficiency_score (0-100)
# Based on energy delivered per second (normalized)
# ============================================
master_df['energy_rate'] = master_df['energy_delivered'] / master_df['duration']

# Normalize to 0-100
min_er = master_df['energy_rate'].min()
max_er = master_df['energy_rate'].max()
master_df['efficiency_score'] = ((master_df['energy_rate'] - min_er) / (max_er - min_er)) * 100
master_df['efficiency_score'] = master_df['efficiency_score'].clip(0, 100).round(2)

# Drop helper column
master_df = master_df.drop(columns=['energy_rate'])

print("\n✅ TARGET 4: efficiency_score")
print(f"   Min : {master_df['efficiency_score'].min():.2f}")
print(f"   Mean: {master_df['efficiency_score'].mean():.2f}")
print(f"   Max : {master_df['efficiency_score'].max():.2f}")

# ============================================
# FINAL PREVIEW
# ============================================
print("\n" + "="*55)
print("🎯 ALL 4 TARGETS CREATED")
print("="*55)
targets = ['stress_label', 'health_score', 'RUL', 'efficiency_score']
print(master_df[['battery_id', 'cycle_number', 'capacity'] + targets].head(8).round(2))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# ✅ Load final dataset
master_df = pd.read_csv("master_df_final.csv")

# ============================================
# DEFINE FEATURES (X) — what model sees
# ============================================
feature_cols = [
    'voltage_mean', 'voltage_min', 'voltage_std',
    'current_mean', 'current_std', 'current_min',
    'temp_mean', 'temp_max', 'temp_std',
    'duration', 'voltage_load_mean',
    'voltage_range', 'temp_rise',
    'power_mean', 'energy_delivered',
    'ambient_temperature', 'cycle_number',
    'capacity_drop_pct'
]

# ============================================
# DEFINE TARGETS (y) — what model predicts
# ============================================
X = master_df[feature_cols]

y_stress     = master_df['stress_label']       # Classification
y_health     = master_df['health_score']       # Regression
y_rul        = master_df['RUL']                # Regression
y_efficiency = master_df['efficiency_score']   # Regression

# ============================================
# SPLIT — 80% train, 20% test
# random_state=42 ensures same split every run
# ============================================
X_train, X_test, \
y_stress_train, y_stress_test, \
y_health_train, y_health_test, \
y_rul_train, y_rul_test, \
y_eff_train, y_eff_test = train_test_split(
    X,
    y_stress, y_health, y_rul, y_efficiency,
    test_size=0.2,
    random_state=42
)

# ============================================
# VERIFY
# ============================================
print("✅ Split Complete!")
print(f"\n📊 Total samples   : {len(X)}")
print(f"🏋️  Training samples: {len(X_train)} ({len(X_train)/len(X)*100:.0f}%)")
print(f"📝 Testing samples : {len(X_test)}  ({len(X_test)/len(X)*100:.0f}%)")

print(f"\n📋 Features used   : {len(feature_cols)}")
print(f"🎯 Targets         : stress_label, health_score, RUL, efficiency_score")

print(f"\n🔍 Stress label distribution in TRAIN:")
print(y_stress_train.value_counts())
print(f"\n🔍 Stress label distribution in TEST:")
print(y_stress_test.value_counts())

print(f"\n🔍 Health score range in TRAIN: {y_health_train.min():.1f} → {y_health_train.max():.1f}")
print(f"🔍 Health score range in TEST : {y_health_test.min():.1f} → {y_health_test.max():.1f}")
```

```python
import pickle
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

# ✅ Load splits
with open("splits.pkl", "rb") as f:
    splits = pickle.load(f)

X_train         = splits['X_train']
X_test          = splits['X_test']
y_stress_train  = splits['y_stress_train']
y_health_train  = splits['y_health_train']
y_rul_train     = splits['y_rul_train']
y_eff_train     = splits['y_eff_train']
feature_cols    = splits['feature_cols']

# ============================================
# MODEL 1: Stress Label Classifier
# ============================================
print("🏋️ Training Model 1: Stress Classifier...")
stress_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1       # use all CPU cores
)
stress_model.fit(X_train, y_stress_train)
print("   ✅ Stress model trained!")

# ============================================
# MODEL 2: Health Score Regressor
# ============================================
print("🏋️ Training Model 2: Health Score Regressor...")
health_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1
)
health_model.fit(X_train, y_health_train)
print("   ✅ Health model trained!")

# ============================================
# MODEL 3: RUL Regressor
# ============================================
print("🏋️ Training Model 3: RUL Regressor...")
rul_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1
)
rul_model.fit(X_train, y_rul_train)
print("   ✅ RUL model trained!")

# ============================================
# MODEL 4: Efficiency Score Regressor
# ============================================
print("🏋️ Training Model 4: Efficiency Score Regressor...")
eff_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1
)
eff_model.fit(X_train, y_eff_train)
print("   ✅ Efficiency model trained!")

# ============================================
# QUICK PEEK: Feature Importance (Top 5)
# ============================================
print("\n" + "="*50)
print("🔍 TOP 5 MOST IMPORTANT FEATURES")
print("="*50)

importances = stress_model.feature_importances_
feat_imp = pd.Series(importances, index=feature_cols).sort_values(ascending=False)
print("\n📊 For Stress Prediction:")
print(feat_imp.head(5).round(3))

importances2 = health_model.feature_importances_
feat_imp2 = pd.Series(importances2, index=feature_cols).sort_values(ascending=False)
print("\n📊 For Health Score:")
print(feat_imp2.head(5).round(3))

print("\n✅ ALL 4 MODELS TRAINED SUCCESSFULLY!")
```

```python
def predict_battery_health(input_data: dict) -> dict:
    """
    Master prediction function for BatteryIQ
    
    Input  : dictionary of 17 sensor features
    Output : dictionary of 4 predictions
    """
    
    # ✅ Convert input to DataFrame in correct feature order
    df = pd.DataFrame([input_data])[feature_cols]
    
    # ✅ Run all 4 models
    stress     = stress_model.predict(df)[0]
    health     = round(float(health_model.predict(df)[0]), 2)
    rul        = max(0, round(float(rul_model.predict(df)[0])))
    efficiency = round(float(eff_model.predict(df)[0]), 2)
    
    # ✅ Add human-friendly explanations
    stress_info = {
        'Low'    : "🟢 Battery is healthy and operating normally.",
        'Medium' : "🟡 Battery showing signs of aging. Monitor closely.",
        'High'   : "🔴 Battery under HIGH stress. Service recommended!"
    }
    
    health_info = (
        "🟢 Excellent"  if health >= 85 else
        "🟡 Good"       if health >= 70 else
        "🟠 Fair"       if health >= 55 else
        "🔴 Poor"
    )
    
    rul_info = (
        "🟢 Long life remaining"    if rul >= 100 else
        "🟡 Moderate life remaining" if rul >= 50  else
        "🟠 Limited life remaining"  if rul >= 20  else
        "🔴 Replace soon!"
    )
    
    efficiency_info = (
        "🟢 Highly efficient"      if efficiency >= 70 else
        "🟡 Moderately efficient"  if efficiency >= 40 else
        "🟠 Low efficiency"        if efficiency >= 20 else
        "🔴 Very low efficiency"
    )
    
    return {
        "stress_label"      : stress,
        "stress_message"    : stress_info[stress],
        "health_score"      : health,
        "health_status"     : health_info,
        "RUL"               : rul,
        "rul_status"        : rul_info,
        "efficiency_score"  : efficiency,
        "efficiency_status" : efficiency_info
    }

print("✅ predict_battery_health() function ready!")

# ============================================
# TEST 1: Healthy Battery (Early Cycle)
# ============================================
healthy_battery = {
    'voltage_mean'      : 3.85,
    'voltage_min'       : 3.20,
    'voltage_std'       : 0.25,
    'current_mean'      : -1.10,
    'current_std'       : 0.20,
    'current_min'       : -1.50,
    'temp_mean'         : 28.0,
    'temp_max'          : 35.0,
    'temp_std'          : 2.5,
    'duration'          : 4200.0,
    'voltage_load_mean' : 3.60,
    'voltage_range'     : 0.65,
    'temp_rise'         : 7.0,
    'power_mean'        : 4.20,
    'energy_delivered'  : 4.90,
    'ambient_temperature': 24.0,
    'cycle_number'      : 10
}

print("\n" + "="*55)
print("🔋 TEST 1: HEALTHY BATTERY (Early Cycle)")
print("="*55)
result1 = predict_battery_health(healthy_battery)
for key, val in result1.items():
    print(f"   {key:22s} → {val}")

# ============================================
# TEST 2: Aging Battery (Mid Cycle)
# ============================================
aging_battery = {
    'voltage_mean'      : 3.55,
    'voltage_min'       : 2.85,
    'voltage_std'       : 0.38,
    'current_mean'      : -1.65,
    'current_std'       : 0.55,
    'current_min'       : -2.10,
    'temp_mean'         : 38.0,
    'temp_max'          : 48.0,
    'temp_std'          : 4.2,
    'duration'          : 3200.0,
    'voltage_load_mean' : 2.90,
    'voltage_range'     : 0.70,
    'temp_rise'         : 10.0,
    'power_mean'        : 5.85,
    'energy_delivered'  : 5.20,
    'ambient_temperature': 24.0,
    'cycle_number'      : 80
}

print("\n" + "="*55)
print("🔋 TEST 2: AGING BATTERY (Mid Cycle)")
print("="*55)
result2 = predict_battery_health(aging_battery)
for key, val in result2.items():
    print(f"   {key:22s} → {val}")

# ============================================
# TEST 3: Critical Battery (Late Cycle)
# ============================================
critical_battery = {
    'voltage_mean'      : 3.20,
    'voltage_min'       : 2.40,
    'voltage_std'       : 0.55,
    'current_mean'      : -2.80,
    'current_std'       : 0.95,
    'current_min'       : -3.50,
    'temp_mean'         : 52.0,
    'temp_max'          : 65.0,
    'temp_std'          : 7.5,
    'duration'          : 2100.0,
    'voltage_load_mean' : 2.20,
    'voltage_range'     : 0.80,
    'temp_rise'         : 13.0,
    'power_mean'        : 8.96,
    'energy_delivered'  : 5.22,
    'ambient_temperature': 24.0,
    'cycle_number'      : 160
}

print("\n" + "="*55)
print("🔋 TEST 3: CRITICAL BATTERY (Late Cycle)")
print("="*55)
result3 = predict_battery_health(critical_battery)
for key, val in result3.items():
    print(f"   {key:22s} → {val}")
```
